###Actividad:

Mostar que se cumple la propiedad conmutativa y asociativa con la operación suma de vectores

```python
import matplotlib.pyplot as plt

def sumaVectores(z1,z2):
   return (z1[0]+z2[0], z1[1]+z2[1])

z1 = (2,1)
z2 = (3,4)

suma = sumaVectores(z1,z2)

#Gráfica la suma
plt.quiver(0, 0, z1[0], z1[1], color='g', units='xy', scale=1)
plt.quiver(z1[0], z1[1],z2[0], z2[1], color='b', units='xy', scale=1)
plt.quiver(0, 0, suma[0], suma[1], color='r', units='xy', scale=1)
plt.axis('scaled')
plt.xlim(-1,5)
plt.ylim(0,5)
plt.show()


def sumaVectores2(z2,z1):
   return (z2[0]+z1[0], z2[1]+z1[1])

z1 = (2,1)
z2 = (3,4)

suma = sumaVectores2(z2,z1)

#Gráfica la suma
plt.quiver(0, 0, z2[0], z2[1], color='b', units='xy', scale=1)
plt.quiver(z2[0], z2[1],z1[0], z1[1], color='g', units='xy', scale=1)
plt.quiver(0, 0, suma[0], suma[1], color='r', units='xy', scale=1)
plt.axis('scaled')
plt.xlim(-1,5)
plt.ylim(0,5)
plt.show()
```

```python
import matplotlib.pyplot as plt

def sumaVectoresa(z1,z2):
    return (z1[0]+z2[0], z1[1]+z2[1])

def sumaVectoresb(z2,z3):
    return (z2[0]+z3[0], z2[1]+z3[1])

def sumaVectoresc(z1,z2,z3):
    return (z1[0]+z2[0]+z3[0], z1[1]+z2[1]+z3[1])

z1 = (1, 2)
z2 = (2, 2)
z3 = (3, 1)
sumaa = sumaVectoresa(z1,z2)
sumab = sumaVectoresb(z2,z3)
sumac = sumaVectoresc(z1,z2,z3)


plt.quiver (0,0, z1[0], z1[1], color='r',units='xy',scale=1)
plt.quiver (z1[0], z1[1], z2[0], z2[1], color='g',units='xy',scale=1)
plt.quiver (sumaa[0], sumaa[1], z3[0], z3[1], color='y',units='xy',scale=1)
plt.quiver (0,0, sumac[0], sumac[1], color='b',units='xy',scale=1)

plt.axis('scaled')
plt.xlim(0,6)
plt.ylim(0,6)
plt.show()
```
