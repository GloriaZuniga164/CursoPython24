##Sean las funciones:

f(x)=\sqrt(-x) g(x)=ln x

Graficar: f o g, g o f, fg


```python
from sympy import symbols, sqrt, ln
from sympy.plotting import plot
def f(x):
   return sqrt(-x)
x = symbols('x')

def g(x):
   return ln(x)
print(f(g(x)))
plot(f(g(x)), line_color='blue')
print(g(f(x)))
plot(g(f(x)), line_color='red');
print(f(x)*g(x))
plot(f(x)*g(x), line_color='green')
```

###Actividad

(P + 5/v²)(V-0.03)=9.7

```python
from sympy import symbols, Eq
from sympy import plot_implicit
p, v = symbols('p v')
plot_implicit(Eq((p + 5/v**2)*(v - 0.03) - 9.7, 0), (p, -5, 7),
              (v, -7, 7));
```

```python
from sympy import symbols, cos, sin, exp, pi
from sympy import plot_parametric
t = symbols('t')
def x(t):
  return cos(t) + 1/2*cos(7*t) + 1/3*sin(17*t)
def y(t):
  return sin(t) + 1/2*sin(7*t) + 1/3*cos(17*t)

plot_parametric((x(t), y(t)), (t, 0, 14*pi));
plot_parametric((x(t), y(t)), (t, -50, 50));
```

Cardioide

r(ϴ) = a(1+cos(ϴ))


```python
from sympy import symbols, cos, sin, exp, pi
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 1000)

r = 5*(1 + np.cos(theta))

plt.polar(theta, r, marker = '+', color=(.2,.4,.5,.3));

plt.polar(theta, r, 'r')
plt.title('Cardioide')
plt.show()
```
